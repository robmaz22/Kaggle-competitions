###3. Preparing data for training

* Removing columns with NaN values

```python
train_df = train_df.dropna(1)
test_df = test_df.dropna(1)
```

* Checking common columns for train and test sets

```python
train_columns = list(train_df.columns)
test_columns = list(test_df.columns)
common_columns = list(set(train_columns) & set(test_columns))
common_columns.append('SalePrice')
```

```python
data = train_df[common_columns]
data.head()
```

* Looking for correlated columns with SalePrice column

```python
correlation threshold = 0.2

numerical_feautures = []
numerical_columns = []

for index, value in (data.corr()['SalePrice'] > correlation threshold).iteritems():
  if value:
    numerical_feautures.append(index)
  numerical_columns.append(index)
```

* Separating categorical columns

```python
categorical_feautures = []

for column in data.columns:
  if column not in numerical_columns:
    categorical_feautures.append(column)
```

```python
numerical_X = train_df[numerical_feautures]
y = numerical_X.pop('SalePrice')
```

* Categorical column encoding

```python
from sklearn.preprocessing import OneHotEncoder

categorical_X = train_df[categorical_feautures]
encoder = OneHotEncoder()
categorical_X = encoder.fit_transform(categorical_X)
categorical_X = pd.DataFrame(categorical_X.toarray())
```

```python
X = pd.concat([numerical_X, categorical_X], axis=1)
```

```python
X.head()
```

* Spliting into train and validation sets

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X,y)
```

###4. Training and evaluating

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

model = LinearRegression()
model.fit(X_train, y_train)

accuracy = model.score(X_val, y_val)
print(f'Model accuracy: {accuracy:.2f}%')

log_val = np.log10(y_val)
log_pred = np.log10(model.predict(X_val))
rmse = np.sqrt(mean_squared_error(log_val, log_pred))
print(f'RMSE: {rmse:.4f}')
```

###5. Predicting prices and creating submission file

```python
numerical_feautures.remove('SalePrice')
n_X_test = test_df[numerical_feautures]
c_X_test = test_df[categorical_feautures]

c_X_test = encoder.transform(c_X_test)
c_X_test = pd.DataFrame(c_X_test.toarray())

X_test = pd.concat([n_X_test, c_X_test], axis=1)

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
submission_dict = {'Id': test_df.Id,
                   'SalePrice' : y_pred}

submission_df = pd.DataFrame(submission_dict)
submission_df.head()
```

```python
submission_df.to_csv('submission.csv', index=False)
print('File created')
```
